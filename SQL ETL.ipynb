```python
import pandas as pd

# Load CSVs
sales_df = pd.read_csv("Edits May 4/output_D.csv")
games_df = pd.read_csv("Edits May 4/games.csv")
platform_df = pd.read_csv("Edits May 4/platform.csv")
publisher_df = pd.read_csv("Edits May 4/publisher.csv")

# Merge to get IDs
sales_df = sales_df.merge(games_df[['title', 'title']], on='title', how='left')
sales_df = sales_df.merge(platform_df[['platform', 'platform']], on='platform', how='left')
sales_df = sales_df.merge(publisher_df[['publisher', 'publisher']], on='publisher', how='left')

# Now rename the columns to match sale_df10
sales_df = sales_df.rename(columns={
    'game_id': 'game_id',
    'platform_id': 'platform_id',
    'publisher_id': 'publisher_id',
    'na_sales': 'NA_Sales',
    'eu_sales': 'EU_Sales',
    'jp_sales': 'JP_Sales',
    'other_sales': 'Other_Sales',
    'global_sales': 'Global_Sales'
})

# Keep only needed columns and drop text columns like title/platform/publisher
final_sales = sales_df[['game_id', 'platform_id', 'publisher_id', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

# Export cleaned CSV for import
final_sales.to_csv("cleaned_sales.csv", index=False)
```

```python

import pandas as pd

# Load CSVs
sales_df = pd.read_csv("demo_IDs/sales.csv")
games_df = pd.read_csv("demo_IDs/games.csv")
platform_df = pd.read_csv("demo_IDs/platform.csv")
publisher_df = pd.read_csv("demo_IDs/publisher.csv")

# Step 1: Merge with games to get game_id
sales_df = sales_df.merge(games_df[['ID', 'Title']], on='Title', how='inner', validate='many_to_many')
sales_df = sales_df.rename(columns={'ID': 'game_id'})

# Step 2: Merge with platform to get platform_id
sales_df = sales_df.merge(platform_df[['ID', 'Platform']], on='Platform', how='inner')
sales_df = sales_df.rename(columns={'ID': 'platform_id'})

# Step 3: Merge with publisher to get publisher_id
sales_df = sales_df.merge(publisher_df[['ID', 'Publisher']], on='Publisher', how='inner', validate='many_to_one')
sales_df = sales_df.rename(columns={'ID': 'publisher_id'})

# Step 4: Keep only the necessary columns
final_sales = sales_df[['game_id', 'platform_id', 'publisher_id', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

final_sales
```

```python
# Step 5: Check for any missing IDs
#missing_ids = final_sales[final_sales.isnull().any(axis=1)]
df = final_sales.drop_duplicates(inplace=True)
df
```
